```python
def intermediateLoop(o, n, v, l):
    if n == o and v != 0:
        l.append("full loop end at " + str(n+1))
    elif n != o and n in l:
        l.append("subloop end at " + str(n+1))
    elif (n+1) in selfLocating:
        l.append("self-locating end at " + str(n+1))
    else:
        if n != -1:
            try:
                l.append(n)
                n = occurList[n]
                v = 1
                intermediateLoop(o, n, v, l)
            except IndexError:
                l.append("out of occur list bounds")
        else:
            l.append("out of digit bounds")
```

```python
def findLoops():
    loopList = []
    for num in range(0, len(occurList)):
        listyList = []
        try:
            intermediateLoop(num, num, 0, listyList)
        except RecursionError:
            listyList.append("long")
        for num in range (0, len(listyList)):
            if type(listyList[num]) == int:
                listyList[num] += 1
        loopList.append(listyList)
    return loopList
```

```python
listOfLoops = findLoops()
```
